```python
import soccerdata as sd
import pandas as pd

seasons = ['2018', '2019', '2020', '2021', '2022', '2023','2024','2025']

all_games = []

for s in seasons:
    understat = sd.Understat('ESP-La Liga', s)
    games = understat.read_schedule()
    games['season'] = s
    all_games.append(games)

df = pd.concat(all_games, ignore_index=True)

print(df.head())
```

Cantidad de goles por partido

```python
df['total_goals'] = df.home_goals + df.away_goals

import matplotlib.pyplot as plt

df['total_goals'].hist(bins=10)
plt.title('Distribucin goles por partido')
plt.xlabel('Goles totales')
plt.ylabel('Frecuencia')
plt.show()
```

Promedio de Goles por partido

```python
df['total_goals'].mean()
```

xG vs goles reales

```python
import seaborn as sns

sns.scatterplot(x=df.home_xg, y=df.home_goals)
plt.title('xG vs goles local')
plt.show()
```

```python
#Correlacin xgoles y home goles
df[['home_xg', 'home_goals']].corr()
```

XG Goles para Visita

```python
df['xg_diff'] = df.home_xg - df.away_xg

sns.histplot(df['xg_diff'], bins=30)
plt.title('Diferencia xG')
plt.show()
```

```python
df = df.dropna(subset=['home_goals', 'away_goals'])
```

```python
import numpy as np
df['result'] = np.where(
    df.home_goals > df.away_goals, 'H',
    np.where(df.home_goals < df.away_goals, 'A', 'D')
)
```

Relacin XG diff vs Ganador de partido.
En este grfico lo que se muestra es cual de los dos equipos genera ms oportunidades de acuerdo a si gana o pierde local. Variable XG_diff est definida por XG local - XG Visita

```python
sns.boxplot(x='result', y='xg_diff', data=df)
plt.show()
```

Ranking de equipos

```python
home_goals = df.groupby('home_team')['home_goals'].mean()
away_goals = df.groupby('away_team')['away_goals'].mean()

team_goals = home_goals.add(away_goals, fill_value=0)

team_goals.sort_values(ascending=False).head(10)
```
